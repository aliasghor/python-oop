# 4. Instance methods

Instance methods are methods that are bound to the instance of a class. These methods are either to access the instance's attribute values or to modify the instance's attribute values.

```python
class Hero:
    # Class variabel
    jumlah_hero = 0

    def __init__(self, name: str, health: int, power: int,  armor: int) -> None:
        # Instance variables
        self.name = name
        self.health = health
        self.power = power
        self.armor = armor

        Hero.jumlah_hero += 1

    # Void function, method tanpa return, dan tanpa argumen
    def siapa(self) -> None:
        print(f"Hero's name: {self.name}")

    # Method dengan argument, tanpa return
    def healthUp(self, healthUp: int) -> None:
        self.health += healthUp

    # Method dengan return
    def getHealth(self) -> int:
        return self.health
```

```python
hero1 = Hero("Sniper", 100, 10, 5)
hero2 = Hero("Mario Bros", 90, 5, 10)
```

```python
hero1.siapa()
```

```python
hero2.siapa()
```

```python
hero2.healthUp(20)
```

```python
print(hero2.getHealth())
```

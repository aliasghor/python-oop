# 2. Constructor `__init__()`

`__init__()` is a [special method](https://docs.python.org/3/glossary.html#term-special-method) that is called implicitly during object creation. It is also used to initalize the object's attribute implicitly through the `self` parameter.

The keyword `self` refers to the instance of the class; therefore, it is mandatory for us to put its keyword for every special method or instance method. Without it, the Python's interpreter cannot access its attributes.

> **For your information**: `self` is not a keyword, rather it is just a naming convention that all Python programmers agree.

```python
class Hero:

    def __init__(self, name: str, health: int, power: int, armor: int) -> None:
        self.name = name
        self.health = health
        self.power = power
        self.armor = armor

    def __repr__(self) -> str:
        attrs = ", ".join(f"{key}='{value}'" if isinstance(value, str) else 
                          f"{key}={value}" for key, value in vars(self).items())
        return f"{type(self).__name__}({attrs})"
```

```python
# Creating an instance of the class Hero
hero1 = Hero("Gerry", 100, 40, 60)
hero2 = Hero("Mogi", 100, 60, 40)
```

```python
# Print the instances attribute
print(hero1)
print(hero2)
```

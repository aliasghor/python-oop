# 3. Class dan Instance variables

## Instance variables

Instance variables are variables that are unique to each instance of a class. These variables are bound to the instance of a class.

## Class variables

Class variables are variables that are shared across all instances of a class. These variables can be bound to both class and instances of a class.

```python
class Hero: # Template
    # Class variable
    jumlah = 0

    def __init__(self, name: str, health: int, power: int, armor: int) -> None:
        # Instance variables
        self.name = name
        self.health = health
        self.power = power
        self.armor = armor

        Hero.jumlah += 1

    def __repr__(self) -> str:
        attrs = ", ".join(f"{key}='{value}'" if isinstance(value, str) else
                         f"{key}={value}" for key, value in self.__dict__.items())
        return f"{type(self).__name__}({attrs})"
```

```python
class Hamster: # Template
    # Class variable
    species = "Syrian Hamster"

    def __init__(self, name: str, age: bool) -> None:
        # Instance variables
        self.name = name
        self.age = age

    def __repr__(self) -> str:
        attrs = ", ".join(f"{key}='{value}'" if isinstance(value, str) else
                         f"{key}={value}" for key, value in self.__dict__.items())
        return f"{type(self).__name__}({attrs})"
```

```python
# Creating an instance of the class Hero
hero1 = Hero("Gerry", 100, 60, 40)
hero2 = Hero("Mogi", 100, 40, 60)
```

```python
# Creating an instance of the class Cat
hamster = Hamster("Gerry", False)
```

```python
print(hero1)
print(hero1.jumlah)
print(hero2)
print(Hero.jumlah)
```

```python
print(hamster)
print(hamster.species)
```
